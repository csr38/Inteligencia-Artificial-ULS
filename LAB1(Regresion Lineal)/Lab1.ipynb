# Procesamiento de datos

```python
#Importamos la librerias

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importamo el dataset
dataset = pd.read_csv('Salary.csv')
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, 1].values
dataset.head()
```

```python
#Repartimos el dataset entre entrenamiento y test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
#Ajustamos la regresion lineal con el conjunto de entrenamiento
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#ParÃ¡metro theta 0
regressor.intercept_
```

```python
#ParÃ¡metro theta 1
regressor.coef_
```

```python
# PredicciÃ³n para el valor minimo y maximo del conjunto de datos de entrenamiento
X_min_max = np.array([X_train.min(), X_train.max()])
y_train_pred = regressor.predict(X_min_max)
```

```python

#Datos del dataset
plt.plot(X_train, y_train, 'o')
#Regresion Lineal
plt.plot(X_min_max, y_train_pred, '-')
#Predicciones con el set de pruebaas
plt.plot(X_test, y_test, 'x')
plt.xlabel('AÃ±os de experiencia')
plt.ylabel('Salario')
plt.show()
```
